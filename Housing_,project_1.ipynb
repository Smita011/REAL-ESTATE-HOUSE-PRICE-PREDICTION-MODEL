REAL ESTATE HOUSE PRICE PREDICTIONS USING LINEAR REGRESSION

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/Housing.csv')
```

```python
data.head()
```

```python
data['mainroad']=data['mainroad'].apply({'yes':1,'no':0}.get)
```

```python
data.head(2)
```

```python
data['guestroom']=data['guestroom'].apply({"yes":1,"no":0}.get)
```

```python
data.head(2)
```

```python
data["basement"]= data["basement"].apply({"yes":1,"no":0}.get)
```

```python
data.head(2)
```

```python
data['hotwaterheating']=data['hotwaterheating'].apply({"yes":1,"no":0}.get)
```

```python
data.head(2)
```

```python
data['airconditioning']=data['airconditioning'].apply({'yes':1,"no":0}.get)
```

```python
data.head(2)
```

```python
data['prefarea']=data['prefarea'].apply({"yes":1,"no":0}.get)
```

```python
data.head(2)
```

```python
data['furnishingstatus']=data['furnishingstatus'].apply({"furnished":1,"semi-furnished":2,"unfurnished":0}.get)
```

```python
data.head(10)
```

```python
# data= data.drop("furnishingstatus",axis=1) to drop any column
```

Dividing into Dependent and Independent Variables in same sequence as given in Dataframe

```python
x=data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']]
y=data['price']
```

Splitting the data into Train_Test_split model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8)
```

creating our SMLM Using Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
regression= LinearRegression()   # with this activatuing our algorithm
```

```python
regression.fit(x_train,y_train)   # used for training
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
# accuracy of SMLM
regression.score(x,y)
```
